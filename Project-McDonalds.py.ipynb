```python
heading=[]
heading.append('Food_type')
heading.append('Food_name')
for i in range(1,13):
    heading.append(site.find_all('th')[i].text)
print(heading)
```

```python
contents=site.find_all('td')
contents
```

```python
values=[]
values.append(site.find('h3').text)
values.append(contents[0].text)
for i in contents[1:13:1]:
    values.append(i.text)
print(values)              
```

```python
al_values=[]
for i in contents:
    al_values.append(i.text)
print(al_values)
```

```python
chunk_size=13
actual_val=[al_values[i:i+chunk_size] for i in range(0,len(al_values),chunk_size)]
actual_val
```

```python
actual_val[0]
```

```python
cat=site.find('h3').text
cat
```

```python
actual_val[0]
```

```python
for i in actual_val:
    i.insert(0,cat)
```

```python
actual_val2=[]
for i in soup1:
    filter1=i
    contents1=filter1.find_all('td')
    whole_val=[]
    for j in contents1:
        whole_val.append(j.text)
        chunk_size1=13
        actual_val1=[al_values[i:i+chunk_size1] for i in range(0,len(al_values),chunk_size1)]
        category=i.find('h3').text
        
    for k in actual_val1:
        k.insert(0,category)
        actual_val2.append(actual_val1)
```

```python
actual_val3=[]
for i in actual_val2:
    for j in i:
        actual_val3.append(j)
print(actual_val3[0:5])
```
